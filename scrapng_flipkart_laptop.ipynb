```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/laptops/pr?sid=6bo,b5g")
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
## names prices and rating

names = "KzDlHZ"

prices = "Nx9bqj _4b5DiR"

rating = "XQDdHH"
```

```python
laptop_div = soup.find_all(class_=names)
price_div= soup.find_all(class_= prices)
rating_div = soup.find_all(class_= rating)
```

```python
laptop_name = []
for i in laptop_div:
    laptop_name.append(i.text)   
laptop_name 
len(laptop_name)
```

```python
laptop_price = []
for i in price_div:
    laptop_price.append(i.text)
laptop_price
len(laptop_price)
```

```python
laptop_rating = []
for i in rating_div:
    laptop_rating.append(i.text)   
laptop_rating
len(laptop_rating)
```

```python
driver.quit
```

```python
import pandas as pd
```

```python
data = {'Name': laptop_name, 'Price': laptop_price,}
df = pd.DataFrame(data)
print(df)
```

```python
# Save data to a CSV file
df.to_csv('scrape_details_of_laptops_from_Flipkart', index=False)
```
